Нарисуйте график функции:

y(x) = k∙cos(x – a) + b

для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(-3*np.pi, 3*np.pi, 201)

k, a, b = np.linspace(-2, 4, num = 3)
k1, a1, b1 = np.linspace(-4, 8, num = 3)
k2, a2, b2 = np.linspace(-3, 3, num = 3)


plt.figure(figsize = (10, 5))
plt.plot(x, k * np.cos(x - a) + b, label='k, a, b')
plt.plot(x, k1 * np.cos(x - a1) + b1, color='g', label='k1, a1, b1')
plt.plot(x, k2 * np.cos(x - a2) + b2, color='r', label='k2, a2, b2')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend(frameon=False)
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-2, 3, 201)
plt.figure(figsize = (5, 8))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,8) 
plt.grid(True)
plt.show()


def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
print (x1, y1)
print (x2, y2)
```
